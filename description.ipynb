# Seminarbeschreibung GitOps


## Technische Voraussetzungen

* Zugriff auf einen GitLab-Server mit der Berechtigung, eigene Repositories anlegen zu können
* Lokale Installation einer IDE, bevorzugt Visual Studio Code
* Zugriff auf eine Kubernetes-Umgebung

## Vorkenntnisse

* Grundlegende Kenntnisse des Software-Entwicklungsprozesses mit DevOps
* Kenntnisse einer Programmiersprache wie Python, Java oder C#
  * Hinweis: Dies ist kein Programmierseminar
*  Bedienung der Web-Frontends von GitLab oder GitHub


## Ablauf

* 8 Unterrichtseinheiten mit jeweils 90 Minuten
* Präsentation
* Diskussion
* Eigenständige Übungen

## Inhalte

* Einführung
    * Was ist GitOps? Gemeinsamkeiten und Abgrenzung zu DevOps
    * Zentrale Prinzipien
        * Git-Repository als Single Source of Truth
        * Infrastructure As Code
        * Pull- bzw. Merge-Requests
        * CI/CD
    * Die Rolle von Containern / Kubernetes
    * Ein einfaches Beispiel-Projekt
* Technische Grundlagen
  * GitFlows mit Pull- / Merge-Requests
  * CI/CD
    * Pipeline und Jobs
    * Umsetzung mit GitLab / GitLab Runners
  * Container / Docker - Bauen von Docker Images
  * Kubernetes
    * Architektur und Arbeitsweise
    * Deployment mit Helm-Charts
* Der GitOps Workflow
  * Definition des Soll-Zustands
  * Custom Resource Definitions (CRDs)
  * Drift Detection - der GitOps Operator
  * Die Reconciliation Loop
* GitOps in der Praxis
  * Vom simplen Programmcode zum GitOps-Projekt
  * Deployment neuer Versionen
  * Änderung des Desired States
  * Umgang mit Fehlersituationen
* Abschluss
    * Wann lohnt sich GitOps?
    * Grenzen von GitOps
    * Weiterführende Ressourcen
    * Q&A
